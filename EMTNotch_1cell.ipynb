```python
DSargs.xdomain  = {'W': [0, 4.0e+5],'Z': [0, 1.0e+6],'Y': [0, 5.0e+5],'S': [0, 6.0e+5]
                  ,'N': [0, 5.0e+5],'D': [0, 5.0e+5],'J': [0, 5.0e+5],'I': [0, 5.0e+5]
                  }

ODE = Vode_ODEsystem(DSargs)
plt.figure(figsize=(6*4,5*2))
n = 20
e = 1.0
ic = {}
for j in DSargs.varspecs.keys():
    ic[j] = (1-e*np.random.uniform(-1,1,n))*fp[2][j] + 0.1*fp[0][j]

for i in range(n):
    ODE.set(ics = {'N': ic['N'][i], 'D': ic['D'][i], 'J': ic['J'][i], 'I': ic['I'][i]
                  ,'W': ic['W'][i], 'Y': ic['Y'][i], 'S': ic['S'][i], 'Z': ic['Z'][i]} )
    traj = ODE.compute('traj')
    pts  = traj.sample(dt=1.0)    
    count = 1
    for i in DSargs.varspecs.keys():
        plt.subplot(2,4,count)
        plt.plot(pts['t'], pts[i], '-', c='c', linewidth=1)
        count += 1

for j in DSargs.varspecs.keys():
    ic[j] = (1-e*np.random.uniform(-1,1,n))*fp[0][j]+ 0.1*fp[2][j]

for i in range(n):
    ODE.set(ics = {'N': ic['N'][i], 'D': ic['D'][i], 'J': ic['J'][i], 'I': ic['I'][i]
                  ,'W': ic['W'][i], 'Y': ic['Y'][i], 'S': ic['S'][i], 'Z': ic['Z'][i]} )
    traj = ODE.compute('traj')
    pts  = traj.sample(dt=1.0)                   
    count = 1
    for i in DSargs.varspecs.keys():
        plt.subplot(2,4,count)
        plt.plot(pts['t'], pts[i], '-', c='c', linewidth=1)
        plt.ylabel(i)
        count += 1

plt.savefig('./figures/TimeDynamics.pdf' , format='pdf', dpi=200)
```
